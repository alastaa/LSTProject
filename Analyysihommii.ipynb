```python
import pandas as pd
```

```python
df = pd.read_csv('fixations0102.txt', sep='\t', index_col=False)
df
```

```python
def aoi_group(area):
    area_type = area[0]
    types = {
        'f': 'face',
        'n': 'nowhere',
        'l': 'logo',
    }
    return types[area_type]
    
```

```python
df['AOI group'] = df['AOI'].map(aoi_group)
df.head()
```

```python
import os

for filename in os.listdir('saccades'):
    df = pd.read_csv('saccades/'+filename, sep='\t', index_col=False)
    df['Saccade Target AOI group'] = df['Saccade Target AOI'].map(aoi_group)
    df = df.drop(['Age', 'Saccade Target AOI', 'Sex', 'Handedness', 'Comments', 'SubjectCategory', 'Trial Category', 'SlideNr', 'Unnamed: 19'], axis=1)
    first_saccades = df[df['CountInTrial'] == 1].reset_index().drop('index', axis=1)
    first_saccades.to_csv('first_saccades/first_'+filename, sep=';', index=False)
```
